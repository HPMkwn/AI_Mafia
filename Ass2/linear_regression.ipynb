```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
xdf = pd.read_csv('./data/Linear_X_Train.csv')      # defined data
ydf = pd.read_csv('./data/Linear_Y_Train.csv')
xt = pd.read_csv('./data/Linear_X_Test.csv')
# print(xdf.head())
# print(ydf.head())

x = xdf.x.values
y = ydf.y.values
print(x)
length = x.shape[0]   #numbera of data
print(length)

itr = 100           #define numbers of iterations
ler = 0.01          #define learning rate
```

```
plt.scatter(x,y,label='Training Data')
plt.show()
```

```
def hypothesis(x,theta):
  return theta[0] + theta[1]*x


def error(X,Y,theta):

  err=0
  for i in range(length):
    hx = hypothesis(X[i],theta)
    err += (hx - Y[i])**2

  return 0.5*err


def gradient(X,Y,theta):
  grad = np.array([0.0,0.0])

  for i in range(length):
    hx = hypothesis(X[i],theta)
    grad[0] += (hx - Y[i])
    grad[1] += (hx - Y[i])*X[i]

  return grad


def gradientDescent(X,Y,itr,ler):
  theta = np.array([+1.0,0.0])
  for i in range(itr):
    grad = gradient(X,Y,theta)
    e = error(X,Y,theta)
    theta[0] = theta[0] - ler * grad[0]
    theta[1] = theta[1] - ler * grad[1]
  return theta
```

```
theta = gradientDescent(x,y,itr,ler)
print(theta)
```

```
plt.scatter(x,y,label='Training Data')
plt.scatter(xt,hypothesis(xt,theta),color='orange',label="Prediction")
plt.show()
```
