```
#define error function

def error(x,y,w,b):
  m = x.shape[0]

  for i in range(m):
    hx = hypothesis(x[i],w,b)
    err += y[i]* np.log2(hx) + (1-y[i])*log2(1-hx)

  return err/m
```
